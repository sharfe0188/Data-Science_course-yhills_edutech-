```python
x = np.array([1,2,3,4])
y = np.array([0,3,5,7])
plt.plot(x,y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title("This is the title")
plt.grid(axis = 'x',color = 'red')
plt.show()
```

```python
#Plot 1
x = np.array([1,2,3,4])
y = np.array([0,3,5,7])
plt.subplot(1,2,2) #row,col,ind
plt.plot(x,y)
plt.title("Plot 1")
plt.show()

#Plot2
x = np.array([1,2,5,9])
y = np.array([0,3,5,9])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Plot 2",color = 'green')
plt.show()

#Plot 3
x = np.array([1,2,3,4])
y = np.array([0,3,5,7])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title("Plot 3")
plt.show()

#Plot 4
x = np.array([1,2,5,9])
y = np.array([0,3,5,9])
plt.subplot(2,4,4)
plt.plot(x,y)
plt.suptitle("This is the main title")
plt.show()
```

```python
x = np.array(['mobiles','grocery','games','vehicles'])
y = np.array([3,8,1,10])

plt.bar(x,y,width = 0.2)
plt.show()
```

```python
x = np.array(['mobiles','grocery','games','vehicles'])
y = np.array([3,8,1,10])

plt.barh(x,y,color = 'red',height = 0.1)
plt.show()
```

```python
y = np.array([25,15,40,20])
myexplode = [0,0,0,0.1]
plt.pie(y,explode = myexplode)
plt.show()
```

```python
y = np.array([25,15,40,20])
mylabels = ['Blue','Orange','Green','Red']
mycolors = ['black','blue','red','hotpink']
plt.pie(y,labels = mylabels,shadow = True,colors = mycolors)
plt.legend(title = 'These are the colors',loc ='upper left',fontsize = 5)
plt.show()
```
